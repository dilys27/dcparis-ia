## EX 1 Visualisation d'un nuage de points et exploration des variables

```python
#1)

import pandas as pd
import seaborn as sns
import matplotlib as mp
import numpy as np

print("lib loaded")
```

```python
#2)

houses = pd.read_csv("./DataSources/house_pricing.csv", sep=",")
print("dataset loaded")
houses.head(10)
```

```python
houses.dtypes
houses.info()
```

```python
#3)

df1 = houses[['LotArea', 'SalePrice']]
df1.plot(x="LotArea",y="SalePrice", kind="scatter")
sns.pairplot(df1)
```

```python
#5)

p1 = df2[['LotFrontage', 'LotArea']]
p1.plot(x="LotFrontage",y="LotArea", kind="scatter")
sns.pairplot(p1)
```

```python
p3 = df3[['LotFrontage', 'LotArea']]
p3.plot(x="LotFrontage",y="LotArea", kind="scatter")
sns.pairplot(p3)
```

## EX2 Visualisation de relations continues

```python
#1)

import matplotlib.pyplot as plt

print("lib loaded")
```

```python
dataset = pd.read_csv("./DataSources/sales_predictions.csv", sep=",")

print("dataset loaded")

dataset.head()
```

```python
dataset.info()
```

```python
#2)

sns.relplot(x='date', y='item_price', ci=None, data=dataset[['item_price', 'date']])
```

On peut voir que ce graphe est illisible car on utilise toutes les donnÃ©es de notre fichier (et il en comporte un nombre important).

```python
#3)

ds = dataset[['item_price', 'date']].sample(50, random_state=1)
sns.relplot(x='date', y='item_price', ci=None, data=ds)
```

Pour corriger le problÃ¨me, il suffit d'utiliser la fonction sample() pour avoir accÃ¨s Ã  un Ã©chantillon des donnÃ©es de notre fichier.

```python
#4)

f, ax = plt.subplots(figsize=(20,6))
sns.relplot(x='date', y='item_price', ci=None, data=ds, height=6, kind='line', aspect=3)

#fig = plt.figure(figsize=(20,6))
#plt.scatter(ds["date"], ds["item_price"])
```

Ce graphe nous permet de mieux visualiser les Ã©carts de valeur entre chaque donnÃ©e.

```python
#5)

ds['date'] = pd.to_datetime(ds['date'])
ds.info()
```

```python
#plt.scatter(ds["date"], ds["item_price"])
sns.relplot(x='date', y='item_price', ci=None, data=ds, height=6, kind='line', aspect=3)
```
