## Formaliser des KPI

#### 1) Se renseigner sur ce qu'est un KPI

Un KPI ou Key Performance Indicator (Indicateur clé de performance) est un élément chiffré permettant l'analyse et la mesure de l'efficacité d'une campagne marketing.

Cela donne à l'entreprise l'idée d'un objectif à atteindre et d'observer la tendance pour réussir cette objectif par l'élaboration ou le réajustement de leur stratégie marketing.

Ces KPIs sont suivis quotidiennement et compilés dans des tableaux de bord, appelés reportings.

#### 9) Quel est l'item le plus commandé ?

```python
popular = dataset.groupby('item_name')

print(popular.groups)
```

```python
dicoQuantity = dict() #Quantité totale de chaque item commandé

for item, group in popular:
    dicoQuantity[item] = group['quantity'].sum()
    print(item)
    print(group)
    print("Quantité totale achetée =", group['quantity'].sum())
    print("--------------------------------------")
```

```python
dicoQuantity
```

L'item le plus commandé (en quantité) est donc le 'Chicken Bowl' avec 761. 

#### 10) Combien d'items ont été commandés au total chez Chipotle ?

```python
print("Au total, chez Chipotle,", sum(dicoQuantity.values()), "items ont été commandés.")
```

#### 11) Combien de chiffre d'affaire a réalisé Chipotle ?

```python
prix = dataset['item_price'].apply(lambda x: x[1:]) #Suppression du 1er caractère ($) de chaque ligne de prix
prix = prix.apply(pd.to_numeric) #Convertion des données en numérique pour les additionner

print("Chipotle a réalisé au total un chiffre d'affaire de", prix.sum(), "$.")
```

#### 12) Quel est le chiffre d'affaire moyen par commande ?

```python
df = dataset.groupby('ID')

print(df.groups)
```

```python
dicoOrder = dict() #Somme totale de chaque commande

for item, group in df:
    dicoOrder[item] = pd.to_numeric(group['item_price'].apply(lambda x: x[1:])).sum()
    print(item)
    print(group)
    print("Somme totale de la commande =", pd.to_numeric(group['item_price'].apply(lambda x: x[1:])).sum())
    print("--------------------------------------")
```

```python
mean = sum(dicoOrder.values()) / len(dicoOrder)

print("Le chiffre d'affaire moyen par commande est de", mean, "$.")
```
